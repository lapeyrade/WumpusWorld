```python
import json
import matplotlib.pyplot as plt
import numpy as np

# 15x15 - 1 Agent, 5 pits, 5 wumpus, 1 gold

# File paths
file_paths = [
    "25_01_03_02_49_53-Prolog-Cupid-Brave.json",
    "25_01_03_02_50_15-Basic-Cupid-Brave.json",
    "25_01_03_02_50_25-BehaviourTree-Cupid-Brave.json",
    "25_01_03_02_50_35-FiniteStateMachine-Cupid-Brave.json",
]

# Load and process data
ai_types = []
cumulated_turn_durations = {}

for path in file_paths:
    with open(path, "r") as f:
        data = json.load(f)
        ai_type = data[0]["aiType"]
        turn_durations = data[0]["turnDuration"]
        cumulated_turn_durations[ai_type] = np.cumsum(turn_durations)
        ai_types.append(ai_type)

# Plot
plt.figure(figsize=(10, 6))
for ai_type, cum_durations in cumulated_turn_durations.items():
    plt.plot(cum_durations, label=ai_type)

plt.yscale('log')
plt.title("Cumulative Turn Durations Across AI Types")
plt.xlabel("Turn Index")
plt.ylabel("Cumulative Duration (in ms)")
plt.legend(title="AI Types")
plt.grid(True)
plt.show()
```
